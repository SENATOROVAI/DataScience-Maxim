```python
def delen(frame):
    """
    Разделяет фрейм данных на кусочки по нужному количеству строк и возвращаает список фреймов
    """

    sp = []
    n = 0
    for st in range(10, 301, 10):
        sp.append(frame[n:st])
        n += 10
    return sp


def delit_pic(lst):
    """
    Приминяет к кажому фрейму из списка алгоритм нахождения выбросов и удаляет их из фрема в списке\n
    Выбросы = наблюдения > Q3 + 1,5*IQR или Q1 – 1,5*IQR\n
    Возвращает очищенные от выбросов списки фреймов
    """
    sp = []
    n = 1.5
    for df in lst:
        Q1 = df.quantile(q=0.25)
        Q3 = df.quantile(q=0.75)
        IQR = df.apply(stats.iqr)
        df = df[~((df < abs(Q1 - n * IQR)) | (df > (Q3 + n * IQR))).any(axis=1)]
        sp.append(df)
    return sp


def cbork(lstr):
    """
    Совмещает все df проходясь по списку фреймов
    """
    df = pd.DataFrame(columns=izm1.columns)
    for fr in lstr:
        df = pd.concat([df, fr])
    return df


def intr(fr):
    """
    Создается пустой df c Nan и к нему добавляются df с выброшеннвми значениями
    Затем все Nan в получ. df замен на 0 для дальнейш их замены на экстраполированные данные по
    функции полученной из разделенного df без выбросов
    """
    cp = pd.DataFrame(columns=izm1.columns, index=izm1.index)
    c = pd.concat([fr, cp], axis=1, sort=True).iloc[:, :4].fillna(0)

    for st in fr:
        f = interpolate.interp1d(
            x=fr[st].index, y=fr[st], fill_value="extrapolate", kind="nearest"
        )
        for i in range(1, 301):
            if c[st][i] == 0:
                c[st][i] = f(i)
    return c


def clean_df_pic(frame):
    """
    Приминение предыдущих функций
    """
    sh1 = delen(frame)
    sh2 = delit_pic(sh1)
    sh3 = cbork(sh2)
    sh4 = intr(sh3)
    return sh4


def dal(df1, df2):
    """
    Вычитание из измеренного результат df фоновых значений
    """
    df = df1 - df2
    return df


def fon(izm_f, fon_f):
    """
    Создание фонового df
    """
    izm1 = clean_df_pic(izm_f)
    fon1 = clean_df_pic(fon_f)
    df0 = dal(izm1, fon1)  # Концентр пыли без пылеподавл. стоковая
    return df0


fonovoe_zn = fon(izm3, fon3)


def effectiv(fo, izme):
    fon_i = clean_df_pic(fo)
    izmer_i = clean_df_pic(izme)
    difference = dal(izmer_i, fon_i)
    df_0 = fonovoe_zn

    effect = (1 - (difference.mean() / df_0.mean())) * 100
    effect = effect.to_frame(name="Effectiveness, %")
    effect = effect.transpose()
    return effect
```
