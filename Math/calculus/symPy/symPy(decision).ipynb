```python
import sympy as sym
from sympy.abc import n
from sympy import sqrt, oo

sym.limit(
    (n * (7 * n) ** (1 / 3) - (81 * n**8 - 1) ** (1 / 4))
    / ((n + 4 * sqrt(n)) * (sqrt(n**2 - 5))),
    n,
    oo,
)
```

```python
import sympy as sym
from sympy.abc import n
from sympy import sqrt, oo

sym.limit((3**n - 2**n) / (3 ** (n - 1) + 2**n), n, oo)
```

```python
import sympy as sym
from sympy.abc import x
from sympy import sqrt, oo

sym.limit((2 * x**2 - 9 * x + 10) / (2 * x - 5), x, 5 / 2)
```
