```python
dates = pd.date_range("2024-02-12", "2024-02-19")  # date_range - список диапазона дат
dates
```

```python
sp = pd.Series(
    [*range(23, 31)], index=dates
)  # вывод какого-то диапазона с индексами датами
sp
```

```python
sp2 = pd.Series([*range(34, 50, 2)], index=dates)  # диапазон с шагом
sp2
```

```python
sp_diff = sp2 - sp  # разница значений по всем индексам
sp_diff
```

```python
# Повторно сичтываем данные, и присваиваем даты в кач индекса
df = pd.read_csv(
    "https://raw.githubusercontent.com/SENATOROVAI/data2-pd/main/Data/goog.csv",
    parse_dates=["Date"],
    index_col="Date",
)
```

- В pandas параметр **parse_dates** используется при чтении данных из различных источников, таких как файлы CSV или Excel, чтобы указать pandas анализировать указанные столбцы как объекты *datetime* в процессе чтения. Этот параметр обычно используется для обеспечения того, чтобы столбцы даты и времени считывались и сохранялись как надлежащие объекты datetime в результирующем фрейме данных.
- **index_col**, index_colHashable, последовательность хэшируемых значений или False, необязательно
Столбец(ы) для использования в качестве меток строк, обозначаемых либо метками столбцов, либо индексами столбцов. Если задана последовательность меток или индексов, для меток строк будет сформирован мультииндекс. Примечание: index_col=False можно использовать, чтобы заставить pandas не использовать первый столбец в качестве индекса, например, когда у вас неправильно сформированный файл с разделителями в конце каждой строки
