```python
dates = pd.date_range('2024-02-12','2024-02-19') # date_range - список диапазона дат 
dates
```

```python
sp = pd.Series([*range(23, 31)], index=dates) # вывод какого-то диапазона с индексами датами  
sp
```

```python
sp2 = pd.Series([*range(34,50,2)], index=dates)  # диапазон с шагом 
sp2
```

```python
sp_diff = sp2 - sp # разница значений по всем индексам 
sp_diff
```

Разница температур на конкретную дату 

```python
sp_diff['2024-02-12'] # разница по конретн индексу 
```
