- `__init__.py` - это автозагрузчик модулей, подключается автоматически, когда в нашем модуле появл слово `import`, отвечает за инициализацию пакетов 

- как только появлятся в `__init__.py` директория превращается в пакет, управляет пространством имен (что позволят избежать конфликта имен), отвечает за точку входа в пакет, т.е. интр. будет выполнять код в `__init__.py`, (файл помойка) куда мы выгружаем все import, обявление перемнных, классов 

модуль это - файл с расширением .py

пакет - набор модулей
