```python
# Сколькими способами можно переставить яблокоо банан и грушу, если в уусловии задачи присуттсв. перестановка/сочетание/размещение можно исспользовать офрмулу и клсссифиц задачу
# Перестановка без повторения
# P = n!
# 0!=1, 1!=1

# P3 = 3!
# n! = 1*2*3*...(n-2)*(n-1)n
# n!= n*(n-1)
# n! = 3 = 3*2*1
```

```python
# Рекурсия - вызов функции самой себя из функции


from itertools import (
    combinations,
    combinations_with_replacement,
    permutations,
    product,
)


def fact(n):
    if n == 0:
        return "test"
    return f"{n} * {fact(n-1)}"


fact(1)
```

```python
it = permutations(["apple", "pear", "banan"])
len(list(it))
```

```python
# составить трёхцветный флаг из лент цветных тканей. Есть четыре цвета лент

colors = ["белый", "жёлтый", "синий", "красный"]

# С2_3 = (3!)/(2!*1!) = 3
```

```python
for i in combinations(colors, 3):  # сочетнаие без потворений
    print(i)
```

```python
# размещение, это порядок для которой порядок важен
for i in permutations(colors, 3):
    print(i)
```

```python
# размещение с повторениями
digits = range(10)
pin_var = product(digits, repeat=4)
for var in pin_var:
    print(var)
```

```python
latters = "ABCD"
code_vars = combinations_with_replacement(
    latters, 2
)  # размещение с повторением
for var in code_vars:
    print(var)
```

Перестановками называют комбинации, состоящие из одних и тех же  различных объектов и отличающиеся только порядком их расположения.

```python
# Правило произведения: если объект а можно выьрать n способамии, и если послка каждого такого вывбора объект b  можно выбрать m споосбами, то выбор пары a ^ b, то можно осущ m*n способыми 
# a и b в указынном порядке 

# Правило суммы - если объект а можно выьрать n способамии, и если послка каждого такого вывбора объект b  можно выбрать m споосбами, то выбор пары a | b, то можно осущ m+n способыми 


Сколькми способами можно пресетавить буквы в слове май 
```
